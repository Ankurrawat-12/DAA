```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
num_elements = []
worst_case_times = []
average_case_times = []
best_case_times = []
algos = ["Selection Sort", "Bubble Sort", "Insertion Sort", "Merge Sort", "Quick Sort", "Heap Sort" , "Binary Search", "Linear Search", "NQueens"]
currentAlgoIndex = 8
algo = algos[currentAlgoIndex]
fileName = algo.replace(" ", "_") + "_Output.txt"
```

```python
with open(fileName, 'r') as file:
    for line in file:
        data = line.strip().split()
        num_elements.append(int(data[0]))
        best_case_times.append(int(data[1]))
        average_case_times.append(int(data[2]))
        worst_case_times.append(int(data[3]))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Assuming you have the following lists
num_elements = [4, 6, 8, 16]
best_case_times = [0.000000, 0.000000, 0.001000, 0.013000]

plt.figure(figsize=(10, 6))
plt.plot(num_elements, best_case_times, label='Best Case', marker='^', color="green")

plt.xlabel('Number of Queens')
plt.ylabel('Time Taken (seconds)')  # Corrected label to seconds
plt.title('NQueens Performance')
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.grid(True)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.6f}'))  # Corrected formatter
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(num_elements, worst_case_times, label='Worst Case', marker='o')
plt.plot(num_elements, average_case_times, label='Average Case', marker='s')
plt.plot(num_elements, best_case_times, label='Best Case', marker='^')

plt.xlabel('Number of elements')
plt.ylabel('Time Taken (milliseconds)')
plt.title(algo + ' Performance')
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.grid(True)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.0f}'))
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:.0f}'))

plt.show()
```
